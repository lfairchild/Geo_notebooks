# Slate Islands Structural Analysis

Here we redo the Slate Islands shattercone analyses of Halls (1979) and Stesky and Halls (1983) in support of the SRM hypothesis. We use shattercone data from Stesky and Halls (1983) in conjunction with newly developed paleomagnetic data. 

**Take the mean impact direction to be D=86.6, I=51.1, alpha_95=4.3, N=11 from breccia dike data (Fairchild et al., 2016)**

```python
def calc_mean_bedding(bed_az, bed_dip):
    pole_trend = []
    pole_plunge = []
    for i in range(len(bed_az)):
        pole_trend.append((bed_az[i]+180)%360)
        pole_plunge.append(90-bed_dip[i])
    poles = ipmag.make_di_block(pole_trend,pole_plunge)
    poles_mean = pmag.fisher_mean(poles)
    mean_bed_az = (poles_mean['dec']+180)%360
    mean_bed_dip = 90 - poles_mean['inc']
    print(poles_mean)
    return (mean_bed_az, mean_bed_dip)
```

```python
def find_intersection(start_pts, end_pts):
    '''
    start_pts: Nx3-matrix containing starting point of N lines
    end_pts: Nx3-matrix containing end point of N lines
    '''
    si = end_pts-start_pts
    ni = si/(np.linalg.norm(si, axis=1)*np.ones((3,1))).transpose()
    nx = ni.transpose()[0]
    ny = ni.transpose()[1]
    nz = ni.transpose()[2]
    SXX = np.sum(nx**2-1)
    SYY = np.sum(ny**2-1)
    SZZ = np.sum(nz**2-1)
    SXY = np.sum(nx*ny)
    SXZ = np.sum(nx*nz)
    SYZ = np.sum(ny*nz)
    S = np.array([[SXX, SXY, SXZ], [SXY, SYY, SYZ], [SXZ, SYZ, SZZ]])
    CX = np.sum(start_pts.transpose()[0]*(nx**2-1) + start_pts.transpose()[1]*(nx*ny) + start_pts.transpose()[2]*(nx*nz))
    CY = np.sum(start_pts.transpose()[0]*(nx*ny)  + start_pts.transpose()[1]*(ny**2-1) + start_pts.transpose()[2]*(ny*nz))
    CZ = np.sum(start_pts.transpose()[0]*(nx*nz)  + start_pts.transpose()[1]*(ny*nz)  + start_pts.transpose()[2]*(nz**2-1));
    C = np.array([[CX], [CY], [CZ]])
    P_intersect = (np.linalg.lstsq(S,C))
    return P_intersect
```
