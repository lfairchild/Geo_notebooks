```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from scipy import special
%config InlineBackend.figure_formats = {'svg',}
%matplotlib inline
```

```python
#dimensionless time
kn = 6.399e-7
temp_initial = 0
temp_surface = 1000
x = np.linspace(0,0.02)
# t = np.linspace(0,60*60)
for t in range(60, 60*20, 60*2):
    if (np.average(temp_initial) + 273) < 846:
        kn = (567.3/(temp_initial+273)-0.062)*1e-6
    else:
        kn = (0.732 - 0.000135*(temp_initial+273))*1e-6
    plt.plot(x, (temp_initial-temp_surface)*special.erf(x/(2*np.sqrt(kn*t)))+temp_surface, label=str(t/60) + ' minutes')
    temp_initial = (temp_initial-temp_surface)*special.erf(x/(2*np.sqrt(kn*t)))+temp_surface
plt.xlabel('meters')
plt.ylabel('temperature')
plt.legend()
plt.show()
```
