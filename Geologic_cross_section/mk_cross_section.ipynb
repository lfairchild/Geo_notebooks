# Making a geologic cross section
This notebook demonstrates the use of ```mk_cx.py```, a Python module for making geologic cross sections. The function requires two data files: 1) a linear elevation profile and 2) georeferenced structural measurements. 

````python
import mk_cx
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
````

````python
mk_cx.make_cross_section('Silver_City_profile.csv', 'Silver_City_str.csv', min_length=-100, contacts={'Freda/Nonesuch contact':1})
````

````python
mk_cx.make_cross_section('Marble_Pt_profile.csv', 'MP_structures.csv', min_length=0,
                   max_length=32000, ymin=150, ylim = 600, extrap_range = 3000, 
                   contacts={'Freda/Nonesuch contact':21914}, 
                   locations={'Mineral Point':(30000, 31700)})
````
